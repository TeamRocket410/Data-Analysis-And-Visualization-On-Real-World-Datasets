## Data Exploration

### 1. Show that how the different prices for all the 14 assets correlate.

```python
#Let us create a correlation plot on the actual different category of prices for all the 14 assets.

plt.figure(figsize=(10,8))
sns.heatmap(df[['Open','High','Low','Close']].corr(), vmin=-1.0, vmax=1.0, annot=True, cmap='coolwarm', linewidths=0.1)
plt.show()
```

### 2. Compare the relationship of the average closing price and average opening price for each of the 14 assets.

```python
#Grouping the data by Asset ID, which will help us in determining the mean values for all the 14 assets.

df_new= df.groupby('Asset_ID').mean()
df_new
```

```python
# Adding the asset names column according to the asset id.

Assets= ['Binance Coin', 'Bitcoin', 'Bitcoin Cash', 'Cardano', 'Dogecoin', 'EOS.IO', 'Ethereum', 'Ethereum Classic', 'IOTA', 'Litecoin','Maker','Monero','Stellar','TRON']
df_new['Asset_Names'] = Assets
df_new
```

```python
# Now we are gonna visualize our results.

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(df_new.Asset_Names, df_new.Open, color='b')
ax2.plot(df_new.Asset_Names,df_new.Close, color='g')

ax1.set_xlabel('Asset_Names', size= 20)
ax1.set_ylabel('Close', size= 20)
ax2.set_ylabel('Open', size= 20)
ax1.set_xticklabels(df_new.Asset_Names, rotation= 'vertical', size= 10)

fig.show()
```

### 3. Determine the two most popoular assets of all the 14 assets through weight percentage wise.

```python
# Arranging the weights of the asset from highest to lowest.

fd.sort_values("Weight", ascending=False)
```

```python
# Let us see the percentage of weights given to each of the assets. 


fd["Weight_percentage"] = (fd["Weight"] / fd["Weight"].sum()) * 100
fd.sort_values("Weight", ascending=False)
```

```python
#Visulalizing our results.

plt.figure(figsize=(9,5))

plt.xlabel("Assets ID")
plt.ylabel("Weight Percentage")

plt.bar(fd.Asset_ID,fd.Weight_percentage)
plt.show()
```
